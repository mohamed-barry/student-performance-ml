Predicting Student Perfomace Using KNN Algorithm, 

```python
# Transforming categorial coloumns to quantitative
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

df = pd.read_csv("/Users/mohamedbarry/Intro to AI/Assignment 3/student-mat.csv")
df = df.drop('Unnamed: 0', axis= 1)

category_columns = list()
for column in df.columns:
    if df[column].dtype == 'object':
        category_columns.append(column)
        
mapping_functions = dict()
for column in category_columns:
    values = df[column].unique()
    mapping_function = dict()
    for value_idx, value in enumerate(values):
        mapping_function[value] = value_idx
    mapping_functions[column] = mapping_function
    
for column in category_columns:
    df[column] = df[column].map(mapping_functions[column])

for column in category_columns:
    df[column] = df[column].astype('category')
df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np

# Assuming df is your DataFrame and is already loaded

# Define independent variables (features) and dependent variable (target)
X = df.drop(columns=['Performance'])  # Independent variables
Y = df['Performance']  # Dependent variable

threshold = 0.1  # Define your threshold for the coefficients here
features = X.columns.tolist()

while True:
    # Split data into training and testing sets
    X_train, X_test, Y_train, Y_test = train_test_split(X[features], Y, test_size=0.2, random_state=23)

    # Initialize and fit the linear regression model
    model = LinearRegression()
    model.fit(X_train, Y_train)

    # Get the coefficients
    coefficients = model.coef_

    # Create a mask for coefficients above the threshold
    important_features = np.abs(coefficients) >= threshold

    # If no coefficients are below the threshold, break the loop
    if not any(~important_features):
        break

    # Otherwise, update the list of features by removing those below the threshold
    features = [feature for feature, important in zip(features, important_features) if important]

    # Update X to only include important features
    X = X[features]

    if len(features) == 0:
        print("All features dropped.")
        break

print(f"Remaining features: {features}")
```

KNN Model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score

# Define independent variables (features) and dependent variable (target)
Y = df['Performance']  # Dependent variable

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=23)

k_values = [3, 5, 7, 9, 11, 13]

# Perform classification with KNN and test different K values
for k in k_values:
    model_KNN = KNeighborsClassifier(n_neighbors=k)
    model_KNN.fit(X_train, Y_train)
    y_pred_KNN = model_KNN.predict(X_test)
    accuracy_KNN = accuracy_score(Y_test, y_pred_KNN)
    f1_KNN = f1_score(Y_test, y_pred_KNN, average='macro')  # Calculate F1 score
    print("KNN Accuracy with K =", k, ":", accuracy_KNN)
    print("KNN F1 Score with K =", k, ":", f1_KNN)
    print()
```

Gaussian Model

```python
from sklearn.naive_bayes import GaussianNB

# Perform classification with Gaussian Naive Bayes
model_Gaussian = GaussianNB()
model_Gaussian.fit(X_train, Y_train)
y_pred_Gaussian = model_Gaussian.predict(X_test)
accuracy_Gaussian = accuracy_score(Y_test, y_pred_Gaussian)
print("Gaussian Naive Bayes Accuracy:", accuracy_Gaussian)
```

SVC Model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score




c_values = [0.1, 1, 10, 100]

# Perform classification with SVC and test different C values
for c in c_values:
    model_SVC = SVC(C=c, kernel="linear")
    model_SVC.fit(X_train, Y_train)
    y_pred_SVC = model_SVC.predict(X_test)
    accuracy_SVC = accuracy_score(Y_test, y_pred_SVC)
    f1_SVC = f1_score(Y_test, y_pred_SVC, average='macro')  # Calculate F1 score
    print("SVC Accuracy with C =", c, ":", accuracy_SVC)
    print("SVC F1 Score with C =", c, ":", f1_SVC)
    print()
```
